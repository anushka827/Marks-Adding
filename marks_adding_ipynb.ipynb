```python
import csv  # Import the csv module to work with CSV files

# Open the CSV file named 'marks.csv' in read mode
file = open ('marks.csv', 'r')

# Create a CSV reader object to read the file
csvreader = csv.reader(file)

# Read and skip the header row (e.g., Roll Num , Marks)
header = next(csvreader)

# Initialize an empty dictionary to store total marks per student
totalMarks = {}

# Iterate through each row in the CSV file
for row in csvreader:
    x = row[0]  # Extract the Roll Num from the first column
    marks = row[1]  # Extract the Marks from the second column

    # If the student is already in the dictionary, add the new marks to their total
    if x in totalMarks:
        totalMarks[x] = totalMarks[x] + marks
    # If the student is not in the dictionary, add them with their current marks
    else:
        totalMarks[x] = marks

# Print the final dictionary containing total marks for each student
print(totalMarks)

```
